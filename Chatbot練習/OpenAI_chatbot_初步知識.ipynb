```python
def get_completion(model="gpt-3.5-turbo"):#選擇gpt-3.5-turbo
    messages = [
    {'role':'system', 'content':'請使用繞口令來跟我做回應。'},
    {'role':'user', 'content':'你好'},
    {'role':'assistant', 'content':'好久不見'},
    {'role':'user', 'content':'好久不見，今天你要做什麼？'}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=1, # 模型輸出的隨機程度
    )
    print(response.choices[0].message.content)

response = get_completion()
```

```python
def get_completion_from_messages(messages, model="gpt-3.5-turbo", temperature=0):
    # call OpenAI chat_completion API
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature, # 模型輸出的隨機程度
    )

    # 回傳模型生成的回應
    return response.choices[0].message.content

messages = [
{'role':'system', 'content':'你就像饒舌高手一樣，請使用繞口令來跟我做回應。'},
{'role':'user', 'content':'你好'},
{'role':'assistant', 'content':'好久不見'},
{'role':'user', 'content':'好久不見，今天你要做什麼？'}]

response = get_completion_from_messages(messages, temperature=1)
print(response)
```

### 點單機器人

```python
def collect_messages(prompt):

    context.append({'role':'user', 'content':f"{prompt}"})

    response = get_completion_from_messages(context)

    context.append({'role':'assistant', 'content':f"{response}"})

    return response
```
